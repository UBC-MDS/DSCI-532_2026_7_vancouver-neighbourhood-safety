```python
#df.loc[30984, "NEIGHBOURHOOD"] = "Fairview"

# replace missing NaN value with 'Fairview' in the 'neighbourhood'
df.loc[(df['HUNDRED_BLOCK'] == "15XX GRANVILLE BRDG") & (df['NEIGHBOURHOOD'].isna()) & (df['X'].notna()) & (df['Y'].notna()), 'NEIGHBOURHOOD'] = 'Fairview'
```

```python
# fill missing values in Neighbourhood with Not-Specified
df['NEIGHBOURHOOD'] = df['NEIGHBOURHOOD'].fillna('Not Specified')

# convert month number to month name
df['MONTH_NAME'] = pd.to_datetime(df['MONTH'], format='%m', errors='coerce').dt.month_name()
```

```python
# feature engineer time of day column

def time_of_day(hour):
    if  4 <= hour <= 11:
        return "Morning"
    elif 12 <= hour <= 17:
        return "Afternoon"
    else:
        return "Evening/Night"
```

```python
# apply the function to create column 'TIME_OF_DAY'
df['TIME_OF_DAY'] = df['HOUR'].apply(time_of_day)
```
